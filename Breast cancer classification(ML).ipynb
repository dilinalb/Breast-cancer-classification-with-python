```python
df = df.drop(columns = ['id','Unnamed: 32'])
```

```python
df['diagnosis'].unique()
```

```python
X = df.drop(columns = 'diagnosis')
```

```python
y = df['diagnosis']
```

```python
encoder = LabelEncoder()
```

```python
encoded_y = encoder.fit_transform(y)
```

```python
encoded_y
```

```python
encoded_ydf = pd.DataFrame(encoded_y, columns = ['diagnosis'])
```

```python
encoded_ydf
```

```python
df2 = X.join(encoded_ydf)
```

```python
df2
```

```python
df2_corr = df2.corr()
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(df2_corr, annot = True, cmap = 'Blues')
plt.savefig('heatmap.png', dpi=300)  # Change the filename and dpi as needed
```

```python
encoded_y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,encoded_y, test_size = 0.3)
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print('Accuracy is: ', accuracy*100) #we will get a higher accuracy
```
